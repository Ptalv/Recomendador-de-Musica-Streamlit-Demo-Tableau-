```python
df = df_spotify.copy()
```

```python

def normalize_genres(x):
    s = str(x).lower().strip()

    if s in ("nan", "none", "null"):
        return "unknown"

    s = re.sub(r"[\[\]']", " ", s)
    s = s.replace(",", " ")
    s = s.replace("-", "")
    s = re.sub(r"[^a-z0-9\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()

    return s if s else "unknown"


df["artist_genres_clean"] = df["artist_genres"].apply(normalize_genres)
df["artist_genres_clean"].value_counts().head(10)

def to_token_list(s):
    if s == "unknown":
        return []
    return s.split()

df["genres_list"] = df["artist_genres_clean"].apply(to_token_list)

from collections import Counter

K = 3
topk_by_artist = (
    df.groupby("artist_name")["genres_list"]
      .apply(lambda lists: [g for g, _ in Counter([t for lst in lists for t in lst]).most_common(K)])
)

artist_to_keygenres = topk_by_artist.apply(lambda xs: " ".join(xs) if xs else "unknown").to_dict()

df["artist_genres_key"] = df["artist_name"].map(artist_to_keygenres).fillna("unknown")


```

```python
df[["artist_name","artist_genres","artist_genres_clean","artist_genres_key"]].head(15)

```

```python
df["explicit"] = df["explicit"].astype(int)
df["album_type"] = df["album_type"].astype(str)
df["artist_genres_key"] = df["artist_genres_key"].astype(str)

num_cols = ["track_popularity", "artist_popularity", "artist_followers", "track_duration_ms_final"]
cat_cols = ["album_type", "explicit"]
text_col = "artist_genres_key"
```

```python
preprocess = ColumnTransformer(
    transformers=[
        ("genres", TfidfVectorizer(min_df=2, ngram_range=(1,2)), text_col),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ("num", StandardScaler(), num_cols),
    ],
    remainder="drop"
)

X_main = preprocess.fit_transform(df)

knn_main = NearestNeighbors(n_neighbors=30, metric="cosine")
knn_main.fit(X_main)
```

```python

def recommend_for_user(liked_track_ids, top_n=10):
    liked_idx = df.index[df["track_id"].isin(liked_track_ids)].tolist()
    if len(liked_idx) == 0:
        raise ValueError("No encontré esos track_id en el dataset.")

    user_vec = X_main[liked_idx].mean(axis=0)
    user_vec = np.asarray(user_vec).reshape(1, -1)

    distances, indices = knn_main.kneighbors(user_vec, n_neighbors=top_n + len(liked_idx))

    rec_idx = [i for i in indices[0] if i not in liked_idx][:top_n]
    rec_dist = [d for i, d in zip(indices[0], distances[0]) if i not in liked_idx][:top_n]

    out = df.loc[rec_idx, ["track_id", "track_name", "artist_name", "artist_genres_key"]].copy()
    out["distance"] = rec_dist
    out["rank"] = range(1, len(out)+1)
    return out.sort_values("distance").reset_index(drop=True)

```

```python
liked = df["track_id"].sample(3, random_state=42).tolist()

print("🎧 Likes (base):")
print(df[df["track_id"].isin(liked)][["track_name","artist_name","artist_genres_key"]])

recs = recommend_for_user(liked, top_n=10)
recs
```

```python

df.to_csv("spotify_tableau_base.csv", index=False)

liked = df["track_id"].sample(3, random_state=42).tolist()
recs = recommend_for_user(liked, top_n=50)
recs.to_csv("spotify_tableau_recommendations.csv", index=False)

print("🚀 CSVs listos para Tableau:")
print("- spotify_tableau_base.csv")
print("- spotify_tableau_recommendations.csv")
```

```python

num_features = ["track_popularity", "artist_popularity", "track_duration_ms_final"]

X_num = df[num_features].fillna(0).values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_num)

pca = PCA(n_components=2, random_state=42)
components = pca.fit_transform(X_scaled)

df["PC1"] = components[:, 0]
df["PC2"] = components[:, 1]

df.to_csv("spotify_for_tableau22.csv", index=False)

plt.scatter(df["PC1"], df["PC2"])
plt.title("Scatter 2D con PCA (datos escalados)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()

print("Listo 🚀 CSV y scatter simples generados.")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

cluster_features = [
    "track_popularity",
    "artist_popularity",
    "artist_followers",
    "track_duration_ms_final",
    "album_total_tracks"
]

Xc = df[cluster_features].fillna(0).values
Xc = StandardScaler().fit_transform(Xc)

# KMeans (cluster para CV)
kmeans = KMeans(n_clusters=5, random_state=42, n_init="auto")
df["cluster"] = kmeans.fit_predict(Xc)

# PCA 2D (solo para visualización)
pca = PCA(n_components=2, random_state=42)
pcs = pca.fit_transform(Xc)
df["PC1"] = pcs[:, 0]
df["PC2"] = pcs[:, 1]

df.to_csv("spotify_tableau_base_plus.csv", index=False)
print("Listo: spotify_tableau_base_plus.csv (base + cluster + PC1/PC2)")
```
